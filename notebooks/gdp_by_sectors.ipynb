# Magyar negyedéves GDP alakulása szektorok szerint 2015-2019

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
PROCESSED_DIR: Path = Path("../data/processed")

file_path = PROCESSED_DIR / "gdp_detailed.csv"
df = pd.read_csv(
    file_path,
    parse_dates=["Dátum"],
    index_col="Dátum",
)

cols=["Mezőgazdaság", "Ipar", "Szolgáltatások"]
ax = df[cols].plot(
    title="Magyar negyedéves GDP alakulása szektorok szerint 2015-2019 [millió forint]",
    style=['o-', 's-', '^-', 'D-'],
    figsize=(10, 6) 
)
ax.legend(bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=3)
plt.tight_layout()
plt.savefig("../docs/figures/gdp_by_sectors.png", dpi=300, bbox_inches='tight')
plt.show()
```
