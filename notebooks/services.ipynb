# Magyar negyedéves GDP alakulása szektorok szerint 2015-2019

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
PROCESSED_DIR: Path = Path("../data/processed")

file_path = PROCESSED_DIR / "services.csv"
df = pd.read_csv(
    file_path,
    parse_dates=["Dátum"],
    index_col="Dátum",
)
df.info()

ax = df.plot(
    title="Szolgáltatások alakulása 2015-2019 [millió forint]",
    style=['o-', 's-', '^-', 'D-', '*-', 'p-', 'h-', 'v-', '<-', '>-'],
    figsize=(10, 6) 
)
ax.legend(bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=3)
plt.tight_layout()
plt.savefig("../docs/figures/services.png", dpi=300, bbox_inches='tight')
plt.show()
```
